```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("fulldata.csv")
data = data.drop(columns = "Unnamed: 0")
data
```

```python
genelist = ['cdh', 'unc', 'slc', 'psap', 'sgpl', 'vsir', 'actb']
orglist = set(data["Organ"])
```

```python
a1 = data.loc[data['Gene']=='psap'].loc[data['Organ']=='ob']
plt.figure(figsize = (10,10))
ax = sns.barplot(x='Mouse', y='Concentration (copies/µL)', hue='Allele', data=a1.sort_values(by="Sample/NTC/Control"), palette="Paired")
plt.xticks(rotation=70)
plt.show()
a1
```

```python
for i in genelist:
    for j in orglist:
        a1 = data.loc[data['Gene']==i].loc[data['Organ']==j]
        ax = sns.barplot(x='Mouse', y='Concentration (copies/µL)', hue='Allele', data=a1.sort_values(by="Sample/NTC/Control"), palette="Paired")
        plt.xticks(rotation=70)
        plt.title(i + " " + j)
        plt.show()
        a1
```
