```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
#import scipy.stats
#import seaborn as sns



def graph(file, organ, gene, trait = "RE", errors = "CISUM", title= organ + gene, y = 'y'):
    
    #plt.style.use('fivethirtyeight')
    #plt.style.use('seaborn-whitegrid')
    csfont = {'fontname':'Arial', 'fontsize':14}
    plt.rcParams['axes.facecolor'] = 'white'
    plt.figure(figsize = [10, 10])
    plt.title(title,**csfont)    
    s = 60
    edgecolors='#000000'
    alpha = 0.7
    #ct_c = 'g'
    #as_c = 'r'
    #as_c = 'b'
    #ms_c = '#27c3f3ff'
#    ms_c = '#89def8ff'
    #as_c = '#057091ff'
    ex_c = 'red'
    ct_c = 'blue'
    
    
    # divide data to mouse groups, cut organ/gene info, take means
    ct = file[file.Group == 'ctr']  
    ex = file[file.Group == 'exp']
    ct = ct[ct.Organ == organ]
    ex = ex[ex.Organ == organ]
    ct = ct[ct.Gene == gene]
    ex = ex[ex.Gene == gene]
    ct = ct.groupby("Mouse").mean().reset_index()
    ex = ex.groupby("Mouse").mean().reset_index()
    #d = [0.1, 0.6, 1.2, 1.7]



    #for trait,n in (['RCSLC',d[0:2]],['RCUNC',d[2:4]]):
    
    #create two bars with sem 
        
    plt.bar(0.1, ct[trait].mean(), color = 'white', edgecolor='black',width=0.5, alpha = 1, zorder=-1)
    plt.errorbar(0.1, ct[trait].mean(), (ct[trait].sem()), c = '#000000', capsize=9)

    #plt.errorbar(0.1, ct[trait].mean(), xerr = 0.1, c = '#000000')
    
    plt.bar(0.8, ex[trait].mean(), color = 'white', edgecolor='black',width=0.5, alpha = 1, zorder=-1)
    plt.errorbar(0.8, ex[trait].mean(), (ex[trait].sem()), c = '#000000', capsize=9)
    #plt.errorbar(0.8, ex[trait].mean(), xerr = 0.1, c = '#000000')
        

        

    #create two csatters
         
    for i in (list(ct['Mouse'])):
        x =(-0.15+(0.45/len(list(ct['Mouse'])))*(list(ct['Mouse']).index(i) + 1))
        y =(ct.loc[ct['Mouse']==i][trait])
        print(i,float(y))
        if pd.isna(float(y)) == False:
            plt.scatter(x, y,s=s,edgecolors=edgecolors, c = ct_c, alpha = 0.7)    
            plt.errorbar(x, y, xerr=0, yerr= y*(ct.loc[ct['Mouse']==i][errors])/1000,c = ct_c, alpha = 0.7)
            plt.annotate(i,(x+0.01,y+0.01))
    
    for i in (list(ex['Mouse'])):
        x =(0.55+(0.45/len(list(ex['Mouse'])))*(list(ex['Mouse']).index(i) + 1))
        y =(ex.loc[ex['Mouse']==i][trait])
        print(i,float(y))
        if pd.isna(float(y)) == False:
            plt.scatter(x, y,s=s,edgecolors=edgecolors, c = ex_c, alpha = 0.7)    
            plt.errorbar(x, y, xerr=0, yerr= y*(ex.loc[ex['Mouse']==i][errors])/1000,c = ex_c, alpha = 0.7) 
            plt.annotate(i,(x+0.01,y+0.01))
    
    plt.text(0.4,-1, s="for stats")
    
    plt.show()
        
        
        #      for i in (list(AS['Mouse'])):
   #             #plt.annotate(i, (d[1],float(file.loc[file['id1']==i][trait])))
    #        plt.scatter([n[1]+0.0155*np.random.rand()], (AS.loc[file['Mouse']==i][trait]),s=s,edgecolors=edgecolors, c = as_c, alpha = 1)

    #plt.xticks(d, ('Контроль Slc29a3', 'UNCdM1R1 Slc29a3', 'Контроль Unc5b','UNCdM1R1 Unc5b'),**csfont, rotation=70)
#    plt
    #plt.yticks(**csfont)
    #plt.ylabel(y,**csfont)
    #plt.savefig(f"{organ}EXP.png", transparent=True, bbox_inches="tight", dpi = 100)
```
